# Matplotlib Tutorial
This notebook covers essential concepts of **Matplotlib**, the most widely used Python library for data visualization.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Line Plot
Let's create a simple line plot using Matplotlib.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label='Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
```

## Scatter Plot
Scatter plots are useful for visualizing relationships between two variables.

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='r', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

## Bar Chart
A bar chart represents data with rectangular bars.

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 8]
plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

## Histogram
Histograms represent the distribution of numerical data.

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

## Subplots
Matplotlib allows creating multiple plots in one figure using `subplots()`.

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(x, y, color='blue', linestyle='--')
ax[0].set_title('Sine Wave')
ax[1].bar(categories, values, color='red')
ax[1].set_title('Bar Chart')
plt.show()
```
